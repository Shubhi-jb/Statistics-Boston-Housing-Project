**TASK 3: Apply Statistical Tests**

1. Levene’s Test (Equality of Variance)Hypothesis: $H_0: \sigma_1^2 = \sigma_2^2$ (Variances are equal).

```python
group1=boston_df[boston_df['CHAS']==1]['MEDV']
group2=boston_df[boston_df['CHAS']==0]['MEDV']
levene_stat=scipy.stats.levene(group1, group2)
print(f"Levene's test p-value: {levene_stat.pvalue}")
```

2. T-test (CHAS vs MEDV)Hypothesis: $H_0: \mu_1 = \mu_0$ (No difference in mean house values).

```python
t_stat,p_val=scipy.stats.ttest_ind(group1, group2, equal_var=False)
print(f"T-test p-value: {p_val}")
```

**Interpretation:** Since the p-value (0.0036) is less than 0.05, we reject the null hypothesis. There is a statistically significant difference in median home values between properties bounded by the Charles River and those that are not. Homes near the river tend to have higher median values, suggesting that proximity to the Charles River is a desirable feature that positively influences property prices.

3. ANOVA (AGE Groups vs MEDV)Hypothesis: $H_0$: All age group means are equal.

```python
low_age=boston_df[boston_df['age_group']=='35 years and younger']['MEDV']
mid_age=boston_df[boston_df['age_group']=='between 35 and 70 years']['MEDV']
high_age=boston_df[boston_df['age_group']=='70 years and older']['MEDV']

f_stat,p_val_anova=scipy.stats.f_oneway(low_age, mid_age, high_age)
print(f"ANOVA p-value: {p_val_anova}")
```

**Interpretation:** Since the p-value (1.71 × 10⁻¹⁵) is far below 0.05, we reject the null hypothesis that all age group means are equal. There is a statistically significant difference in median home values across the three age groups. Specifically, newer homes (built for younger populations) tend to command higher median values, indicating that the age profile of a neighbourhood's housing stock has a meaningful impact on property prices.

4. Pearson Correlation (NOX vs INDUS)Hypothesis: $H_0$: No correlation exists.

```python
corr,p_val_corr=scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
print(f"Pearson correlation coefficient: {corr}, p-value: {p_val_corr}")
```

**Interpretation:** The Pearson correlation coefficient of 0.76 indicates a strong positive relationship between nitric oxide concentrations (NOX) and the proportion of non-retail business acres (INDUS). The p-value is effectively 0, confirming this relationship is highly statistically significant. This means towns with greater industrial activity tend to have considerably higher air pollution levels — a finding relevant to urban planning and public health decisions.

5. Regression (DIS on MEDV)

```python
X=boston_df['DIS']
y=boston_df['MEDV']
X=sm.add_constant(X)
model=sm.OLS(y,X).fit()
print(model.summary())
```

**Interpretation:** The OLS regression model shows that the distance to employment centres (DIS) has a statistically significant effect on median home values (MEDV), with a p-value of 0.000. For every additional unit of distance from employment centres, the median home value increases by approximately **1.09 units**. However, the R² value of 0.062 indicates that distance alone explains only 6.2% of the variation in home prices, suggesting that while the relationship is real, many other factors also drive property values in Boston.
